### Definir los Scores

```python
def dotProduct(Q, Doc):
    Q = np.log10(Q+1)
    Doc =  np.log10(Doc+1)
    return np.dot(Q, Doc)

def dotProduct2(Q, Doc):
    return np.dot(Q, Doc)

def cosine(Q, Doc):
    Q = np.log10(Q+1)
    Q_nom = np.linalg.norm(Q)
    nqi = Q/Q_nom
    
    Doc =  np.log10(Doc+1)
    Doc_nom = np.linalg.norm(Doc)
    ndi = Doc/Doc_nom
    return np.dot (nqi,ndi)

def cosine2(Q, Doc):
    Q_nom = np.linalg.norm(Q)
    nqi = Q/Q_nom
    Doc_nom = np.linalg.norm(Doc)
    ndi = Doc/Doc_nom
    return np.dot (nqi,ndi)

```

### Ejercicio 1

```python
def retrieval1(Query, Collection, func_score):
    temp = []
    num = 0
    for doc in Collection:
        num += 1
        score = func_score(Query, doc)
        name = "Doc " + str(num)
        temp2 = [name, score]
        temp.append(temp2)
    return sorted(temp, key=lambda x: x[1],reverse=True)

collection = [
    np.array([15, 5, 20, 25]),
    np.array([30, 0, 22, 0])
]   
query = np.array([115,10,2,0])


```

```python
# query = np.log10(query+1)
# collection = np.log10(collection+1)
retrieval1(query, collection, dotProduct)
```

```python
retrieval1(query, collection, cosine)
```
